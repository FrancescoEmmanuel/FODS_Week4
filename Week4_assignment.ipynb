```python
def clean_column(value):
    try:
        pattern = r"'name':\s+'(.*?)'"
        match = re.findall(pattern, value)
        return ', '.join(match)
    except (TypeError, AttributeError):
        return ''
```

```python
movieDf['belongs_to_collection'] = movieDf['belongs_to_collection'].apply(clean_column)
```
